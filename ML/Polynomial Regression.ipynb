```python
import numpy as np
import matplotlib.pyplot as mp
import pandas as pd

#fetching data
data = pd.read_csv("Position_Salaries.csv",encoding='utf-8')
inp = data.iloc[:,1:2].values
out= data.iloc[:,-1].values

print(out)

```

```python
from sklearn.linear_model import LinearRegression
#LinearRegression

lin_reg1 = LinearRegression()
lin_reg1.fit(inp,out)


#Polynomial Features

from sklearn.preprocessing import PolynomialFeatures

poly_reg = PolynomialFeatures(degree=4)

inp_poly = poly_reg.fit_transform(inp)

poly_reg.fit(inp_poly,out)

regr = LinearRegression()
regr.fit(inp_poly,out)

#Plotting Linear Regression
mp.scatter(inp,out,color = 'red')
mp.plot(inp,lin_reg1.predict(inp),color ='blue')
mp.title("Salary vs Experience(Linear Regression)")
mp.xlabel("Experience")
mp.ylabel("Salary")
mp.show()

#plotting Polynomial Regression
inp_grid = np.arange(min(inp),max(inp),0.1)
inp_grid = inp_grid.reshape(len(inp_grid),1)
mp.scatter(inp,out,color = 'red')
mp.plot(inp_grid,regr.predict(poly_reg.fit_transform(inp_grid)),color ='blue')
mp.title("Salary vs Experience(Polynomial Regression)")
mp.xlabel("Experience")
mp.ylabel("Salary")
mp.show()


```
