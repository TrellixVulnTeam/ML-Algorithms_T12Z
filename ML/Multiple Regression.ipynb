```python
import numpy as np


import matplotlib.pyplot as mp

import pandas as pd

#fetching data
data = pd.read_csv("50_Startups.csv",encoding='utf-8')
inp = data.iloc[:,:-1].values
out= data.iloc[:,-1].values




from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.compose import ColumnTransformer

'''

from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan,strategy='mean')
imputer = imputer.fit(inp[:,1:3])
inp[:,1:3]=imputer.transform(inp[:,1:3])
'''


#encode categorical data

transformer = ColumnTransformer([("State",OneHotEncoder(),[3])],remainder = 'passthrough')
inp = transformer.fit_transform(inp)    

#Avoiding dmmy variable trap
inp = inp[:,1:]



'''
temp = LabelEncoder()
temp = temp.fit_transform(out)
out = temp
'''

#splitting data set into train and test

from sklearn.model_selection import train_test_split

inp_train,inp_test,out_train,out_test = train_test_split(inp,out,test_size = 0.2, random_state = 0)

'''
from sklearn.preprocessing import StandardScaler
SC = StandardScaler()
inp_train = SC.fit_transform(inp_train)
inp_test = SC.transform(inp_test)
'''
```

```python
#multiple regression

from sklearn.linear_model import LinearRegression

regressor = LinearRegression()
regressor.fit(inp_train,out_train)

out_pred = regressor.predict(inp_test)

print(out_pred)
```

```python
#using Backward Elimination

import statsmodels.api as stats

inp = np.append(arr = np.ones((50,1),int), values = inp, axis =1)

inp_opt = np.array(inp[:,[0,1,2,3,4,5]],dtype=float)
opt_regressor = stats.OLS(endog=out,exog=inp_opt).fit()
opt_regressor.summary()


inp_opt = np.array(inp[:,[0,1,3,4,5]],dtype=float)
opt_regressor = stats.OLS(endog=out,exog=inp_opt).fit()
opt_regressor.summary()


inp_opt = np.array(inp[:,[0,3,4,5]],dtype=float)
opt_regressor = stats.OLS(endog=out,exog=inp_opt).fit()
opt_regressor.summary()


inp_opt = np.array(inp[:,[0,3,5]],dtype=float)
opt_regressor = stats.OLS(endog=out,exog=inp_opt).fit()
opt_regressor.summary()


inp_opt = np.array(inp[:,[0,3]],dtype=float)
opt_regressor = stats.OLS(endog=out,exog=inp_opt).fit()
opt_regressor.summary()






```
